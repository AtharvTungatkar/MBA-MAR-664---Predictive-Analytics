```python
import numpy as np
import pandas as pd
```

```python
true = np.arange(1,11,1)
np.random.seed(1031)
pred = np.arange(1,11,1) + np.random.randint(-1,2,10)
data = pd.DataFrame({'true':true,
              'pred':pred})
data
```

```python
(
    data
        .assign(e=data.true-data.pred)
        .sum()
)
```

```python
(
    data
        .assign(e=data.true-data.pred,
               ae=np.abs(data.true-data.pred),
               se=(data.true-data.pred)**2,
               )
)
```

```python
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_error
```

```python
root_mean_squared_error(data.true,data.pred)
```

```python
mean_absolute_error(data.true,data.pred)
```

```python
r2_score(data.true,data.pred)
```

```python
np.random.seed(1031)
true = np.random.choice(a = [0,1],size = 10, replace = True)
pred_prob = np.random.random_sample(size = 10)
data = pd.DataFrame({'true':true,
                     'pred_prob':pred_prob})
data
```

```python
data['pred'] = (data['pred_prob']>0.5).astype('int32')
data
```

```python
ct = pd.crosstab(data.true, data.pred,rownames = ['true'], colnames=['pred'])
TN = ct.iloc[0,0]
TP = ct.iloc[1,1]
FP = ct.iloc[0,1]
FN = ct.iloc[1,0]
ct
accuracy = (TP + TN)/len(data)
precision = TP/(TP + FP)
sensitivity_or_recall = TP/(TP + FN)
specificity = TN/(TN+FP)
pd.Series(data = [accuracy, precision, sensitivity_or_recall, specificity],
          index=['Accuracy','Precision', 'Sensitivity or Recall', 'Specificity'])
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(data.true, data.pred)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
accuracy = accuracy_score(data.true, data.pred)
precision = precision_score(data.true, data.pred)
sensitivity_or_recall = recall_score(data.true, data.pred)
auc = roc_auc_score(data.true, data.pred_prob)

pd.Series(data = [accuracy, precision, sensitivity_or_recall, specificity, auc],
          index=['Accuracy','Precision', 'Sensitivity or Recall', 'Specificity', 'Area Under the Curve'])
```
