```python
ufo=pd.read_csv('ufo.csv')
ufo
```

```python
ufo['reported_date_time'] = pd.to_datetime(ufo.reported_date_time, format = 'mixed')
ufo['year'] = ufo.reported_date_time.dt.year
ufo['month'] = ufo.reported_date_time.dt.month
ufo['day'] = ufo.reported_date_time.dt.day
ufo['weekday'] = ufo.reported_date_time.dt.day_name()
ufo['hour'] = ufo.reported_date_time.dt.hour
ufo['minute'] = ufo.reported_date_time.dt.minute
ufo['second'] = ufo.reported_date_time.dt.second
ufo['bright'] = ufo.summary.str.contains(pat = 'bright')
ufo['craft'] = ufo.summary.str.contains(pat = 'craft')
ufo['light'] = ufo.summary.str.contains(pat = 'light')
ufo.head()
```

```python
np.random.seed(1031)
gender = np.random.choice(['Male','Female'],size = 10,replace = True)
age_group = np.random.choice(['18-29','30-39','40-49','50-59','60 or over'],size = 10,replace=True)
marital_status = np.random.choice(['married','married','widowed','divorced','separated','never married','never married'],size = 10,replace = True)
ethnicity = np.random.choice(['White','Black','Native','Asian','Hispanic','Other'],size = 10,replace = True)
zip_code = np.random.randint(10000,10055,10)
cat_demo = pd.DataFrame({'gender':gender, 
             'age_group':age_group,
             'marital_status': marital_status,
             'ethnicity': ethnicity,
             'zip_code': zip_code})
cat_demo
```

```python
cat_demo.gender.value_counts()
```

```python
pd.get_dummies(cat_demo.gender,drop_first=True)
```

```python
pd.get_dummies(cat_demo.age_group,drop_first=False)
```

```python
pd.get_dummies(cat_demo.marital_status,drop_first=True)
```

```python
marital_counts=counts = cat_demo.marital_status.value_counts()
sparse_categories = marital_counts[counts<2].index
cat_demo['marital_status'] = cat_demo.marital_status.replace(sparse_categories,'other')
cat_demo
```

```python
scaler=StandardScaler()
```

```python
chipotle_scaled=scaler.fit_transform(fastfood[['chipotle']])
```

```python
from sklearn.feature_selection import VarianceThreshold
low_var = VarianceThreshold(threshold=0.1)
fastfood_likert = fastfood.loc[:,'health1':'shake_shack']
fastfood_likert_reduced = low_var.fit_transform(fastfood_likert)
fastfood_likert_reduced
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_normal = MinMaxScaler(feature_range=(0,1))
min_max_normal.fit_transform(fastfood.loc[:,'health1':'health3'])
```

This uses an operation similar to logarithm

```python
from sklearn.preprocessing import PowerTransformer
box_cox = PowerTransformer('box-cox')
box_cox_duration = box_cox.fit_transform(fastfood.loc[:,['duration']])
box_cox_duration.reshape(2000,)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree = 2, include_bias=False, interaction_only=False)
poly_features.fit_transform(fastfood.loc[:,'health1':'health2'])
```
