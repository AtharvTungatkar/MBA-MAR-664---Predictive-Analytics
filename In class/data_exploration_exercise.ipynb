## 1) Setup
**Question:** Import the required libraries: `numpy`, `pandas`, and `matplotlib.pyplot`.

```python
# Your code here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2) Reading Data
**Question:**
A. Use `pd.read_csv()` to load `tech_companies.csv` into a DataFrame called `companies`.
B. Print the first five rows.

```python
# A. Read CSV
companies=pd.read_csv('tech_companies.csv')

# B. Head of data
companies.head()
```

## 3) Inspecting Data
**Question:**
A. Print the structure of the DataFrame using `.info()`.
B. Display the data types of each column.
C. Generate summary statistics with `.describe(include='all')`.
D. Check the shape, columns, and index of the DataFrame.

```python
# A. info()
print(companies.info())

# B. dtypes
print(companies.dtypes)

# C. describe
print(companies.describe(include='all'))

# D. shape, columns, index
print(companies.shape)
print(companies.columns)
print(companies.index)
```

## 4) Subsetting Data
**Question:**
A. Select only the `Company` column.
B. Select both `Revenue` and `Profit_Margin`.
C. Slice the first 10 rows.
D. Select rows where `Profit_Margin > 25`.

```python
# A. one column
companies['Company']

# B. two columns
companies[['Revenue','Profit_Margin']]

# C. first 10 rows
companies.iloc[0:10,:]

# D. filter condition
companies[companies.Profit_Margin>25]
```

## 5) Editing Data
**Question:**
A. Create a new column `Revenue_Millions` by dividing `Revenue` by 1e6.
B. Rename `Profit_Margin` to `ProfitPct`.
C. Drop the `Employees` column.

```python
# A. new column
companies['Revenue_Millions']=companies.Revenue/1000000

# B. rename column
companies.rename(columns={'Profit_Margin':'ProfitPct'},inplace=True)

# C. drop column
companies.drop(labels='Employees',axis=1)
```

## 6) Summarizing Data
**Question:**
A. Compute the average profit margin.
B. Find the maximum revenue and the company that earned it.
C. Group companies by `Industry` and compute their mean revenue.

```python
# A. average profit margin
average_profit_margin=np.mean(companies.Profit_Margin)

# B. max revenue and company
companies[companies.Revenue==companies.Revenue.max()][['Company','Revenue']]

# C. group by industry
companies.groupby(by='Industry',as_index=False).agg({'Revenue':'mean'})
```

## 7) Visualization
**Question:**
A. Create a histogram of `ProfitPct`.
B. Create a scatter plot of `Revenue` vs. `ProfitPct`.
C. Add axis labels and a title.

```python
# A. histogram
companies.ProfitPct.hist()

# B. scatter plot
plt.plot(companies.Revenue,companies.ProfitPct)

# C. labels and title
plt.title('Scatter Plot')
plt.xlabel('Revenue')
plt.ylabel('ProfitPct')
```
