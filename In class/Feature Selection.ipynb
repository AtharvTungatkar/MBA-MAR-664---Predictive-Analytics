```python
import pandas as pd
import numpy as np
wine = pd.read_csv('Datasets/winequality-white.csv',sep=';')
wine.columns = wine.columns.str.replace(' ', '_')
```

```python
from sklearn.model_selection import train_test_split
X = wine.drop(['quality'], axis=1)
y = wine.quality
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3, 
                                                    random_state=1031)
```

```python
wine_train = X_train.copy()
wine_train['quality'] = y_train
wine_train.corr().round(2)
```

```python
# wine_train.corr().round(2)['quality'] # to see only relevant predictors
wine_train.corr().quality.round(2).sort_values(ascending=True)
```

```python
# wine_train.corr().round(2).drop('quality', axis = 1) # to see only relevant predictors
wine_train.corr().round(2).drop('quality', axis = 1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
sns.heatmap(wine_train.corr(), 
            annot=True, 
            fmt=".2f", 
            cmap="RdYlGn",
            vmin=-1, 
            vmax=1, 
            square=True)
plt.tight_layout()
plt.show()
```

```python
# VIF is used to detect multicollinearity
from statsmodels.formula.api import ols
reg = ols('quality~fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol',
    data=wine_train)
reg.fit().summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.api import add_constant
X_with_const = add_constant(X_train)

vif_data = pd.DataFrame()
vif_data["feature"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i)
                   for i in range(X_with_const.shape[1])]

vif_data.sort_values('VIF',ascending=False)
```

```python
# Forward Pass

from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
sfs = SequentialFeatureSelector(LinearRegression(),
                                  k_features='best',       
                                  forward=True,
                                  floating=False,          
                                  scoring='r2',
                                  cv=5)
sfs = sfs.fit(X_train, y_train)
pd.DataFrame(sfs.get_metric_dict()).T.loc[:,['feature_names','avg_score']].sort_values('avg_score', ascending = False)
```

```python
pd.DataFrame(sfs.get_metric_dict()).T.loc[:,['feature_names','avg_score']].sort_values(by='avg_score',ascending=False)
```

```python
# Backward Pass

from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
sfs = SequentialFeatureSelector(LinearRegression(),
          k_features='best',       
          forward=False,
          floating=False,          
          scoring='r2',
          cv=5)
sfs = sfs.fit(X_train, y_train)
pd.DataFrame(sfs.get_metric_dict()).T.loc[:,['feature_names','avg_score']].sort_values('avg_score', ascending = False)
```

```python
sfs.k_feature_names_
```

```python
# Stepwise

from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
sfs = SequentialFeatureSelector(LinearRegression(),
          k_features='best',       
          forward=True,
          floating=True,           
          scoring='r2',
          cv=5)
sfs = sfs.fit(X_train, y_train)
pd.DataFrame(sfs.get_metric_dict()).T.loc[:,['feature_names','avg_score']].sort_values('avg_score', ascending = False)
```

```python
sfs.k_feature_names_
```

```python
# Best Subset Selection using all possible combinations of features

from mlxtend.feature_selection import ExhaustiveFeatureSelector
from sklearn.linear_model import LinearRegression
efs = ExhaustiveFeatureSelector(LinearRegression(),
                                min_features=1,
                                max_features = 11,
                                print_progress=True,
                                scoring = 'r2',
                                cv=5)
efs = efs.fit(X_train, y_train)
pd.DataFrame(efs.get_metric_dict()).T.loc[:,['feature_names','avg_score']].sort_values('avg_score', ascending = False)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit(X_train)
X_train_scaled = scale.transform(X_train)
X_test_scaled = scale.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet, ElasticNetCV
lasso = ElasticNetCV(l1_ratio=1,   # Lasso if l1 ratio is 1 if ridge then put it to 0
                     cv=5,
                     random_state=1031)
lasso.fit(X_train_scaled, y_train)
best_alpha = lasso.alpha_
print('The best alpha is', best_alpha)
```

```python
lasso.coef_
```

```python
pd.DataFrame(lasso.coef_,columns=['lasso coefficients']).set_index(X_train.columns)
```

```python
alphas = lasso.alphas_
mse_path = lasso.mse_path_
mean_mse = np.mean(mse_path, axis=1)
std_mse = np.std(mse_path, axis=1)
```

```python
plt.figure(figsize=(8, 6))
feature_names = X_train.columns
coef_paths = []
for alpha in alphas:
    lasso_temp = ElasticNet(l1_ratio=1, alpha=alpha, random_state=1031, max_iter=2000)
    lasso_temp.fit(X_train_scaled, y_train)
    coef_paths.append(lasso_temp.coef_)
coef_paths = np.array(coef_paths)
# Plot of coefficient paths
for i, feature in enumerate(feature_names):
    plt.plot(np.log10(alphas), coef_paths[:, i], label=feature, linewidth=1.5)    
plt.axvline(x=np.log10(best_alpha), color='red', linestyle='--', alpha=0.7, linewidth=2)
plt.xlabel('log(alpha)')
plt.ylabel('Coefficient Values')
plt.title('Lasso: Coefficient Paths')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()
```

```python
# Plot of number of non-zero coefficients vs alpha
non_zero_counts = np.sum(coef_paths != 0, axis=1)
plt.plot(np.log10(alphas), non_zero_counts, 'sienna', linewidth=3)
plt.axvline(x=np.log10(best_alpha), color='red', linestyle='--', alpha=0.7, linewidth=2)
plt.xlabel('log(alpha)')
plt.ylabel('Number of Non-Zero Coefficients')
plt.title('Lasso: Feature Selection')
plt.show()
```
