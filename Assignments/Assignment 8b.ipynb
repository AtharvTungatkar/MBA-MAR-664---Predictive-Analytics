```python
import numpy as np
import pandas as pd

data=pd.read_csv('Datasets/spotify_data.csv')
```

```python
data.drop(labels=['id', 'performer', 'song', 'genre', 'key'],axis=1,inplace=True)

y=data['rating']
X=data.drop(labels='rating',axis=1)

y_binned = pd.qcut(data.rating, q = 20)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1031, stratify=y_binned)
```

Question 1

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import root_mean_squared_error

base_tree = DecisionTreeRegressor(max_depth=10, random_state=1031)


bag_model = BaggingRegressor(
    estimator=base_tree,
    n_estimators=100,
    oob_score=True,
    bootstrap=True,
    random_state=1031,
    n_jobs=-1
)

bag_model.fit(X_train, y_train)

y_pred_bag = bag_model.predict(X_train)

rmse_train_bagging = root_mean_squared_error(y_train, y_pred_bag)
rmse_train_bagging.round(4)
```

Question 2

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.linear_model import LinearRegression

base_lr = LinearRegression()

bag_lr_model = BaggingRegressor(
    estimator=base_lr,
    n_estimators=100,
    bootstrap=True,
    random_state=1031,
    n_jobs=-1
)

bag_lr_model.fit(X_train, y_train)

y_pred_bag_lr = bag_lr_model.predict(X_train)

# ---------------------------------------
# 5. Compute TRAIN RMSE
# ---------------------------------------
rmse_train_bag_lr = root_mean_squared_error(y_train, y_pred_bag_lr)
rmse_train_bag_lr.round(4)
```

Question 3

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=100,       
    bootstrap=True,
    random_state=1031,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_train)

rmse_train_rf = root_mean_squared_error(y_train, y_pred_rf)
rmse_train_rf.round(4)
```
