Question 2

```python
from statsmodels.formula.api import logit

log_model=logit('attrition ~ gender',data=train)
model2=log_model.fit()

model2.summary()
```

```python
logit_female=(np.exp(model2.params[0])/(1+np.exp(model2.params[0])))
logit_male=(np.exp(model2.params.sum())/(1+np.exp(model2.params.sum())))

abs_diff=np.abs(logit_male-logit_female)*100
abs_diff.round(4)
```
