```python
import numpy as np
import pandas as pd

data=pd.read_csv('Datasets/airline_satisfaction.csv')
```

Question 1

```python
data.drop(labels=['age', 'gender', 'frequent_flyer'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X=data.drop(labels='satisfaction',axis=1)
y=data[['satisfaction']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,stratify=y,random_state=1031)
```

```python
np.median(X_train['flight_distance'])
```

Question 2

```python
combined=pd.concat([X_train,y_train],axis=1)
combined.corr()[['satisfaction']]
```

Question 3

```python
X_train.corr()
```

Question 4

```python
from statsmodels.formula.api import ols

all_columns='+'.join(combined.columns.difference(['satisfaction']))
formula='satisfaction ~ '+all_columns
model=ols(formula=formula,data=combined).fit()
```

```python
model.summary()
```

Question 5

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.api import add_constant
X_with_const=add_constant(X_train)

vif_data=pd.DataFrame()
vif_data['feature']=X_with_const.columns
vif_data['VIF']=[variance_inflation_factor(X_with_const.values,i) for i in range(X_with_const.shape[1])]

vif_data[vif_data['VIF']>5]
```
