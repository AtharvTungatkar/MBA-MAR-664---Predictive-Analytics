```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('restaurant_rating.csv')
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size=0.2, random_state=617)

train_mean = train['rating'].mean()
test_mean = test['rating'].mean()

abs_diff = round(abs(train_mean - test_mean), 4)

print("Average rating (Train):", train_mean)
print("Average rating (Test):", test_mean)
print("Absolute difference:", abs_diff)
```

```python
bins = pd.cut(data["rating"], bins=20)
train, test = train_test_split(data, test_size=0.2, random_state=617, stratify=bins)
train_mean = train["rating"].mean()
test_mean = test["rating"].mean()
abs_diff = round(abs(train_mean - test_mean), 4)
print(abs_diff)
```
