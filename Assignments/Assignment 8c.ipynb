```python
import numpy as np
import pandas as pd

data=pd.read_csv('Datasets/spotify_data.csv')
```

```python
data.drop(labels=['id', 'performer', 'song', 'genre', 'key'],axis=1,inplace=True)

y=data['rating']
X=data.drop(labels='rating',axis=1)

y_binned = pd.qcut(data.rating, q = 20)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1031, stratify=y_binned)
```

Question 1

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import root_mean_squared_error

base_tree = DecisionTreeRegressor()

ada_model = AdaBoostRegressor(
    estimator=base_tree,
    n_estimators=100
)

ada_model.fit(X_train, y_train)

y_pred_ada = ada_model.predict(X_train)

# ---------------------------------------
# 5. Compute RMSE
# ---------------------------------------
rmse_train_ada = root_mean_squared_error(y_train, y_pred_ada)
rmse_train_ada.round(4)
```

Question 2

```python
from sklearn.ensemble import GradientBoostingRegressor


gb_model = GradientBoostingRegressor(n_estimators=100)

gb_model.fit(X_train, y_train)

y_pred_gb = gb_model.predict(X_train)

rmse_train_gb = root_mean_squared_error(y_train, y_pred_gb)
rmse_train_gb.round(4)
```

Question 3

```python
sgb_model = GradientBoostingRegressor(
    n_estimators=100,        
    subsample=0.6,      
    max_features=0.4,  
)

sgb_model.fit(X_train, y_train)

y_pred_sgb = sgb_model.predict(X_train)


rmse_train_sgb = root_mean_squared_error(y_train, y_pred_sgb)
rmse_train_sgb.round(4)
```
